```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('df_anscombe.csv')
```

```python
#Reusmenes estadísticos de cada grupo
df.groupby('group').describe()
```

```python
groups = df['group'].unique()
n_groups = len(groups)
fig, axes = plt.subplots(nrows=1, ncols=n_groups, figsize=(20, 5), sharey=True)

for i, group in enumerate(groups):
    # Seleccionamos los datos del grupo actual
    group_data = df[df['group'] == group]
    ax = axes[i]
    # Graficamos los datos de cada grupo en un scatter plot
    ax.scatter(group_data['x'], group_data['y'])
    ax.set_title(f'Group {group}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.suptitle('Gráficas de dispersión de cada grupo')
plt.tight_layout()
plt.show()
```

```python
# 4. Graficar box-plots de cada grupo
fig, axes = plt.subplots(nrows=1, ncols=n_groups, figsize=(20, 5), sharey=True)

for i, group in enumerate(groups):
    group_data = df[df['group'] == group]
    ax = axes[i]
    # Graficamos el box-plot para cada grupo
    ax.boxplot(group_data['y'])
    ax.set_title(f'Box-plot Group {group}')
    ax.set_xlabel('Group')
    ax.set_ylabel('Values')

plt.suptitle('Box-plot de cada grupo')
plt.tight_layout()
plt.show()
```

```python
# 4. Graficar box-plots de cada grupo
fig, axes = plt.subplots(nrows=1, ncols=n_groups, figsize=(20, 5), sharey=True)

for i, group in enumerate(groups):
    group_data = df[df['group'] == group]
    ax = axes[i]
    # Graficamos el box-plot para cada grupo
    ax.boxplot(group_data['x'])
    ax.set_title(f'Box-plot Group {group}')
    ax.set_xlabel('Group')
    ax.set_ylabel('Values')

plt.suptitle('Box-plot de cada grupo')
plt.tight_layout()
plt.show()
```
